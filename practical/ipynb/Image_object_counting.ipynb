<font color="#D35400" size='6px'><b>Image Object Counting</b></font>

<font color="#D35400" size='4px'><b>The Counting System</b></font>

The system operates in a rigorous counting pipeline designed to identify, separate, and tally specific objects within a scene. It combines Vision-Language Models for discovery with **SAM 3** for pixel-precise segmentation.

<font color="#117A65" size='4px'><b>Stage I: Scene Understanding & Class Discovery</b></font>
Before counting begins, the system must determine "what" is in the image.

**Automatic Discovery:** The image is analyzed by a Vision-Language Model to generate a list of target objects.
* <font color="#884EA0"><i>LLM Mode:</i></font> Uses **Async Parallel Processing** to analyze multiple crops of the image simultaneously, detecting complex or subtle objects (e.g., "specific brand logo", "vintage lamp").
* <font color="#884EA0"><i>BLIP Mode:</i></font> Uses local captioning models to find standard nouns (e.g., "car", "person").

**NLP Refinement:** The raw text descriptions are parsed using **SpaCy**. Abstract concepts (e.g., "view", "atmosphere") are filtered out, and synonyms are merged to create a clean list of physical target nouns.

<font color="#117A65" size='4px'><b>Stage II: SAM 3 Instance Segmentation</b></font>
The clean noun list serves as the input prompt for the segmentation engine.

**Text-Prompted Inference:** Each discovered noun (e.g., "apple") is fed into the **SAM 3 (Segment Anything Model 3)** as a text prompt.

**Mask Generation:** SAM 3 predicts binary **Spatial Masks** for every instance of that object. Unlike semantic segmentation (which groups all "apples" into one blob), SAM 3 distinguishes between "Apple 1", "Apple 2", and "Apple 3."

<font color="#117A65" size='4px'><b>Stage III: Filtering & Deduplication</b></font>
Raw model outputs can contain noise or overlapping duplicates.

**Confidence Thresholding:** Masks with low confidence scores (below `SAM3_CONF_IMAGE_THRESHOLD`) are discarded to ensure quality.

**Non-Maximum Suppression (NMS):** If two masks overlap significantly (IoU > Threshold), the system keeps only the higher-confidence one, preventing double-counting of the same object.

<font color="#117A65" size='4px'><b>Stage IV: Visualization & Reporting</b></font>
The final step constructs the analytical output.

**Count Aggregation:** The system tallies the valid masks for each class (e.g., "Person: 5", "Car: 3").

**Visual Overlay:**
* <font color="#884EA0"><i>Bounding Boxes:</i></font> Draws a precise box around each object with its ID.
* <font color="#884EA0"><i>Mask Overlay:</i></font> Applies a semi-transparent color mask to show the exact pixel coverage.
* <font color="#884EA0"><i>Legend:</i></font> Displays the final count for every detected class.

<font color="#D35400" size='6px'>Execution2: BLIP Automatic Detection</font>

This cell runs the Standard Automatic Mode. By using <font color="#D35400" size='4px'>args.mode='blip'</font>, employ a locally hosted Vision-Language Model to "look" at image and figure out what's inside without needing external APIs.

The function <font color="#D35400"><i>get_classes_blip</i></font> orchestrates a smart scanning process:

<font color="#D35400" size='4px'><i>Smart Cropping:</i></font> To ensure don't miss small details, the image is sliced into multiple strategic crops (zoom-ins) using <font color="#D35400"><i>visualization.get_image_crops</i></font>. This allows the AI to focus on specific regions rather than just the cluttered whole.

<font color="#D35400" size='4px'><i>Local Captioning:</i></font> Runs the <font color="#c5885f"><i>BlipForConditionalGeneration</i></font> model on every single crop. It generates a descriptive sentence for each section,  creating a textual map of the image content.

<font color="#D35400" size='4px'><i>Noun Extraction:</i></font> Finally, all these captions are stitched together. then uses a standard NLP library (SpaCy) to parse this text and extract the specific nouns (e.g., "car", "dog"), creating a clean target list for SAM3 to segment.

<font color="#3c983c" size='6px'>Execution3: LLM-Enhanced Reasoning</font>

This cell activates the LLM Mode. By simply setting <font color="#3c983c" size='4px'>args.mode='llm'</font>, bypass the standard local models and instead use a Multimodal LLM to get a much deeper understanding of the scene.

This function runs on a fast Async Pipeline:

<font color="#CB4335" size='4px'><i>Asyncio Parallelism:</i></font> unlike BLIP (which works sequentially), the function <font color="#CB4335"><i>get_classes_llm</i></font> chops the image into multiple detailed crops using <font color="#CB4335" size='4px'><i>visualization.get_image_crops*</i></font>. It then sends API requests for all those crops at the exact same time, analyzing the whole image in parallel rather than waiting.

<font color="#CB4335" size='4px'><i>Visual Reasoning:</i></font> Instead of a generic "caption", instruct the model to focus strictly on describing physical objects. This allows the system to spot subtle or complex items that standard BLIP models usually miss.

<font color="#CB4335" size='4px'><i>Intelligent Parsing:</i></font> The raw descriptions from the vision model are collected and sent to a second "Parser" LLM. This step, managed by <font color="#CB4335"><i>utils.search_engine_clean_nouns</i></font>, cleans up the text and extracts a neat list of target nouns that act as precise prompts for SAM3.

In this cell, switch to *Geometric Mode*. Instead of asking the AI to "guess" what to count, we provide specific spatial hints (points or bounding boxes).

When <font color="#2E86C1">*task_type="geometric"*</font>, the pipeline skips the NLP class generation entirely. It directly feeds your input points to SAM3 to force it to pay attention to specific regions.

<font color="#2E86C1" size='4px'>*Input:*</font> A list of coordinates ```(e.g., [[x, y], [x, y]]).```

<font color="#2E86C1" size='4px'>*Process:*</font> The script normalizes these points relative to the image size and passes them to <font color="#2E86C1">*sam3_processor*</font> as foreground prompts.
